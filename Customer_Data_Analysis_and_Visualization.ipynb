```python
import pandas as pd
df = pd.read_csv("/content/customer_shopping_behavior.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Category'].nunique()
```

```python
null_rows = df[df.isnull().any(axis=1)].head(5)
print(null_rows)
```

```python
median_review_rating_per_category = df.groupby('Category')['Review Rating'].transform('median')
df['Review Rating'].fillna(median_review_rating_per_category, inplace=True)
median_per_category = df.groupby('Category', as_index=False)['Review Rating'].median()
print(median_per_category)
print("----------------------------------")
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
df.head()
```

```python
labels = ['Young_Adult', 'Adult', 'Middle_Age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df.head()
```

```python
df['frequency_of_purchases'].value_counts().sort_index()
```

```python
frequency_mapping = {
    'Annually':365,
    'Monthly':30,
    'Weekly':7,
    'Fortnightly':15,
    'Bi-Weekly':14,
    'Quarterly':90,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df.head()
```

```python
# Checking is 'discount applied' and 'promo code used' both columns are same
(df['discount_applied'] == df['promo_code_used']).all()
```
